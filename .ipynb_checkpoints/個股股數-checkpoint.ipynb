# 獲取個股發行股數

```python
smy=pd.read_excel("C:/Users/黃柏倫/Desktop/test2.xlsx",header=0)
```

```python
smy.head()
```

```python
codelist=smy.代號
codelist[0]
```

```python
i=0
t=0
while i <len(smy.index):
    
    for j in range(i,len(smy.index)):
        
        try:
            co_id=codelist[j]

            payload={

            'encodeURIComponent':'1',
            'step':'1',
            'firstin':'1',
            'off': '1',
            'keyword4':'', 
            'code1': '',
            'TYPEK2':'' ,
            'checkbtn':'', 
            'queryName': co_id,
            'inpuType': co_id,
            'TYPEK': 'all',
            'co_id': co_id

            }

            res=requests.post("http://mops.twse.com.tw/mops/web/ajax_t05st03",data=payload)

            res.encoding='utf-8'

            soup=bs(res.text)

            dfs=pd.read_html(soup.prettify('utf-8'),encoding='utf-8')

            smy.loc[j,'股數']=int(int(dfs[1][dfs[1].loc[:,0]=='已發行普通股數或TDR原股發行股數'].iloc[0,2].split("股")[0].replace(',','')))
            print(f'第{j+1}個好了')
            time.sleep(5)
            if j ==len(smy.index):
                i=1000
        except:
            print(f'第{j+1}個爆了')
            if i==j:
                t+=1
            if t>1:
                t=0
                i=j+1
                print(f'放棄第{j+1}個了')
                break
            i=j
            time.sleep(15)
            break
```
