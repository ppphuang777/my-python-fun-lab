# API 

```python
def nor(t):
    t-= t.min()
    t =t/t.max()
    return t
```

```python
dfc=nor(np.array(df['收盤價']))
dfv=nor(np.array(df['成交值']))
```

```python
a=[]
b=[]
y=[]
for i in range(len(df.index)-10):
    a.append(dfc[i:i+10])
    b.append(dfv[i:i+10])
    y.append(df.漲跌平[i+10])
a=np.array(a).reshape(len(x),10,1)
b=np.array(b).reshape(len(y),10,1)
```

```python
b[0]
```

```python
from keras.utils import np_utils
```

```python
a_train=a[:10000]
a_test=a[10000:]
b_train=b[:10000]
b_test=b[10000:]
yc = np_utils.to_categorical(y, 3)
y_train=yc[:10000]
y_test=yc[10000:]
```

```python
y_train
```

```python
from keras.models import Sequential
from keras.layers import Dense, Embedding
from keras.layers import LSTM
```

```python
from keras.models import Model
from keras.layers import Input
from keras.layers import concatenate, add
```

```python
f_1 = LSTM(40, input_shape=(10,1))
g_1 = LSTM(40, input_shape=(10,1))

h_1 = Dense(3, activation='softmax')
```

```python
a0 = Input(shape=(10,1))
b0 = Input(shape=(10,1))
a1 = f_1(a0)

b1 = g_1(b0)


u = concatenate([a1,b1])

y = h_1(u)
```

```python
model3 = Model([a0,b0], y)
model3.summary()
```

```python
model3.compile(loss='binary_crossentropy',optimizer="adam",metrics=['acc'])#,validation)
```

```python
model3.fit([a_train,b_train], y_train,batch_size=1000,epochs=5)
```

```python
model3.evaluate([a_test,b_test], y_test)
```
