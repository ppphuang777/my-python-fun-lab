```python
def get(s,cd,i):
    global vc,vp,ex
    if (list(cd[(cd['履約價']==s)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'])==[]):        
        vc=s+50
        vp=s-50
    else:
        vc=s+ex
        vp=s-ex
    c=float(cd[(cd['履約價']==vc)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'].iloc[0])
    p=float(cd[(cd['履約價']==vp)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'].iloc[1])
    return p+c

def getpc(s,cd,i):
    global vc,vp,ex
    if (list(cd[(cd['履約價']==s+25)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'])==[]):        
        vc=int(round(s/100)*100)+ex
    else:
        vc=s+25+ex      
    if (list(cd[(cd['履約價']==s-25)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'])==[]):        
        vp=int(round(s/100)*100)-ex
    else:
        vp=s-25-ex
    c=float(cd[(cd['履約價']==vc)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'].iloc[0])
    p=float(cd[(cd['履約價']==vp)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'].iloc[1])
    
    return p+c
def buy(s,cd,i):
    global vc,vp,point,ex
    if (list(cd[(cd['履約價']==s)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'])==[]):        
        vc=s+50
        vp=s-50        
    else:
        vc=s+ex
        vp=s-ex
    c=float(cd[(cd['履約價']==vc)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'].iloc[0])
    p=float(cd[(cd['履約價']==vp)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'].iloc[1])
    
    point=max((int(df.loc[i,'收盤價'])-vc),(vp-(int(df.loc[i,'收盤價']))),0)
#    print(f"賠{point}點",end=',')
    return point+p+c

def buypc(s,cd,i):
    global vc,vp,point,ex
    if (list(cd[(cd['履約價']==s+25)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'])==[]):        
        vc=int(round(s/100)*100)+ex
    else:
        vc=s+25+ex
    if (list(cd[(cd['履約價']==s-25)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['開盤價'])==[]):        
        vp=int(round(s/100)*100)-ex
    else:
        vp=s-25-ex
    c=float(cd[(cd['履約價']==vc)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['收盤價'].iloc[0])
    p=float(cd[(cd['履約價']==vp)&(cd['交易日期']==df.loc[i,'交易日期'])&(cd['交易時段']==df.loc[i,'交易時段'])]['收盤價'].iloc[1])
    point=max((int(df.loc[i,'收盤價'])-vc),(vp-(int(df.loc[i,'收盤價']))),0)
#    print(f"賠{point}點",end=',')
    return point+p+c
```
