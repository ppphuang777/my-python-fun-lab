```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dg = pd.read_excel('C:/Users/黃柏倫/Desktop/bingo500.xlsx')
```

```python
df=dg[:]
dh=dg[-5:]
dh.index=range(0,5,1)
```

```python
n=np.array([])
for i in range(1,50):
    n=np.insert(n,i-1,len(df[df[i]>0].index))
```

```python
mu,sig=n.mean(),n.std()
```

```python
c=[]
for nn in n:
    if nn>mu+sig:
        c.append('r')
    elif nn<mu-sig:
        c.append('orange')
    else:
        c.append('g')
```

```python
plt.figure(figsize=(20,5))
plt.bar(range(1,50),n,color=c)
plt.xlabel("Number")
plt.ylabel("Times")
plt.grid(True)
plt.xticks(np.arange(1,50,1))
plt.yticks(np.arange(0,100,10))
for x,y in zip(range(1,50),n):
    plt.text(x,y+0.05,s=f'{int(y)}',ha='center')
```
